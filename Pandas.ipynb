### Group data and summarize it

```python
df.groupby('Entity').min() # max
```

### Manipulate data frame

Remove a column

```python
df = df.drop('Code', axis=1)
df
```

Define the index (i.e. row label) to be the year

```python
df.index = df['Year']
df
```

Remove a column

```python
df = df.drop('Year', axis=1)
df
```

```python
df = df.pivot(columns='Entity', values='Electricity consumption')
df
```

```python
df.dropna()
```

```python
df.fillna(0)
```

```python
df.fillna(method='bfill')
```

```python
df.describe()
```

```python
df['Germany'].plot()
```

```python
df['Germany'].max()
```

```python
small_df = df.iloc[:, :20]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,7))
small_df.plot(legend=False, ax=ax)
ax.legend(bbox_to_anchor=(1.1, 0.9))
```

## Plotting with Seaborn

```python
import seaborn as sns
```

### Boxplot

```python
sns.boxplot(data=df)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10))
sns.boxplot(data=small_df, ax=ax)
ax.set_xticklabels(small_df.columns, rotation=90);
```

### Heatmap

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10))
sns.heatmap(data=small_df.transpose(), ax=ax, cmap='Blues')
```
